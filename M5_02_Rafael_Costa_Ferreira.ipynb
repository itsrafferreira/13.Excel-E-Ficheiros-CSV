**2) Continuar com o Excel**
* Criar uma lista que contenha os seguintes dados (nome do país e medalhas de ouro, prata e bronze):
```
USA, 46, 12, 5
China, 38, 20, 7
UK, 29, 7, 7
Russia, 22, 10, 9
South Korea, 13, 3, 2
Germany, 11, 7, 4
```
* Adicionar essa lista à folha de olimpíadas que criámos no exercício anterior 
* Listar por filas e colunas o conteúdo
* Mover todo o intervalo de dados uma fila para baixo (pesquisar na documentação como mover intervalos)
* Adicionar umas cabeceiras na primeira linha (que acaba de ficar livre) que sejam “País”, “Ouros”, “Pratas”, “Bronzes” 
* Guardar e comprovar as alterações no excel 

````python
# Continuação do código anterior
sheet = workbook["olimpíadas"]
data = [
    ["País", "Ouros", "Pratas", "Bronzes"],
    ["USA", 46, 12, 5],
    ["China", 38, 20, 7],
    ["UK", 29, 7, 7],
    ["Russia", 22, 10, 9],
    ["South Korea", 13, 3, 2],
    ["Germany", 11, 7, 4]
]
for row in data:
    sheet.append(row)

# Mover todo o intervalo de dados uma fila para baixo
sheet.move_range("A1:D7", rows=1, cols=0)

# Guardar e comprovar as alterações no excel
workbook.save("02_Excel_data.xlsx")
````

**3)  Continuar com o Excel, centrando-se na livraria openpyxl.styles (a qual proporciona estilo às nossas folhas de Excel)**
* Colocar as cabeceiras que acabámos de adicionar a negrito (bold) 
* Guardar e comprovar as alterações no excel

Pista: embora haja várias formas de o fazer, o mais comum é definir o estilo que deseja (negrito) e logo aplicá-lo a todas as células que queira (iterar pela primeira fila da nossa folha)


Doc: https://openpyxl.readthedocs.io/en/stable/styles.html

````python
import openpyxl
from openpyxl.styles import Font

workbook = openpyxl.load_workbook("02_Excel_data.xlsx")
sheet = workbook["olimpíadas"]

#Colocar as cabeceiras a negrito
bold_font = Font(bold=True)
for cell in sheet[1]:
    cell.font = bold_font

#Guardar e comprovar as alterações
workbook.save("02_Excel_data.xlsx")
````

**4) Continuar com o Excel. Criar uma nova coluna que seja o somatório de todas as medalhas conseguidas por cada país**

````python
#Criar uma nova coluna que seja o somatório de todas as medalhas conseguidas por cada país
for row in sheet.iter_rows(min_row=2, max_row=sheet.max_row, min_col=2, max_col=4):
    total_medals = sum([int(cell.value) for cell in row if cell.value is not None and str(cell.value).isdigit()])
    sheet.cell(row=row[0].row, column=5, value=total_medals)

workbook.save("02_Excel_data.xlsx")
````

**5)  Continuar com o Excel. Criar um gráfico de barras (openpyxl.chart.BarChart()) o mais parecido ao que se mostra na imagem:**

<img src="./02_Excel_data.png"  />

````python
import openpyxl
from openpyxl.chart import BarChart, Reference

wb = openpyxl.load_workbook('02_Excel_data.xlsx')
ws = wb.active

data = Reference(ws, min_col=5, min_row=2, max_col=5, max_row=7)
categories = Reference(ws, min_col=1, min_row=2, max_row=7)

chart = BarChart()
chart.add_data(data)
chart.set_categories(categories)

ws.add_chart(chart, "E10")
````

## CSV

**Vamos fazer um exercício prático de *Data Mining*, onde iremos limpar e filtrar informação com um csv que contém dados de veículos**

Trabalhar com o ficheiro **02_CSV_data.csv**

**6) Observar os dados do csv e fazer as seguintes tarefas:**

* Temos 9 colunas, as 8 primeiras contêm dados numéricos com os quais podemos trabalhar, criar uma lista para cada uma destas colunas (mpg, cylinders, etc.). Não é necessário criar lista para *name*
* Percorrer os dados do csv adequadamente e armazenar os dados de cada coluna em cada uma das listas que criou anteriormente.
* Comprovar que se guardou nas listas a informação correspondente (com prints)

````python
import csv

#Criar lista para cada coluna
mpg = []
cylinders = []
displacement = []
horsepower = []
weight = []
acceleration = []
model_year = []
origin = []

#Armazenar os dados do csv de cada coluna em cada lista
with open('02_CSV_data.csv', 'r') as file:
    csv_reader = csv.reader(file)
    next(csv_reader)  # Skip header
    for row in csv_reader:
        mpg.append(float(row[0]))
        cylinders.append(int(row[1]))
        displacement.append(float(row[2]))
        horsepower.append(float(row[3]))
        weight.append(float(row[4]))
        acceleration.append(float(row[5]))
        model_year.append(int(row[6]))
        origin.append(int(row[7]))

print(mpg)
print(cylinders)
print(displacement)
print(horsepower)
print(weight)
print(acceleration)
print(model_year)
print(origin)
````

**8) Para terminar vamos mostrar alguns resultados. Para isso iremos utilizar a livraria *statistics*.**

* Mostrar para cada uma das listas, quantas observações tem (quantos dados tem). 
* Mostrar o mínimo e o máximo de cada lista
* Mostrar a média, mediana e desvio padrão de cada lista (utilizando statistics)

````python
import csv
import statistics as stats

mpg = []
cylinders = []
displacement = []
horsepower = []
weight = []
acceleration = []
model_year = []
origin = []

with open('02_CSV_data.csv', 'r') as file:
    csv_reader = csv.reader(file)
    next(csv_reader)  # Skip header
    for row in csv_reader:
        mpg.append(float(row[0]))
        cylinders.append(int(row[1]))
        displacement.append(float(row[2]))
        horsepower.append(float(row[3]))
        weight.append(float(row[4]))
        acceleration.append(float(row[5]))
        model_year.append(int(row[6]))
        origin.append(int(row[7]))

mpg.pop(0)
cylinders.pop(0)
displacement.pop(0)
horsepower.pop(0)
weight.pop(0)
acceleration.pop(0)
model_year.pop(0)
origin.pop(0)

#Mostrar para cada uma das listas, quantas observações tem 
mpg = list(map(float, mpg))
cylinders = list(map(int, cylinders))
displacement = list(map(float, displacement))
horsepower = list(map(float, horsepower))
weight = list(map(float, weight))
acceleration = list(map(float, acceleration))
model_year = list(map(int, model_year))
origin = list(map(int, origin))

print(len(mpg))
print(len(cylinders))
print(len(displacement))
print(len(horsepower))
print(len(weight))
print(len(acceleration))
print(len(model_year))
print(len(origin))

#Mostrar o mínimo e o máximo de cada lista
print(min(mpg), max(mpg))
print(min(cylinders), max(cylinders))
print(min(displacement), max(displacement))
print(min(horsepower), max(horsepower))
print(min(weight), max(weight))
print(min(acceleration), max(acceleration))
print(min(model_year), max(model_year))
print(min(origin), max(origin))

#Mostrar a média, mediana e desvio padrão de cada lista (utilizando statistics)
print(stats.mean(mpg), stats.median(mpg), stats.stdev(mpg))
print(stats.mean(cylinders), stats.median(cylinders), stats.stdev(cylinders))
print(stats.mean(displacement), stats.median(displacement), stats.stdev(displacement))
print(stats.mean(horsepower), stats.median(horsepower), stats.stdev(horsepower))
print(stats.mean(weight), stats.median(weight), stats.stdev(weight))
print(stats.mean(acceleration), stats.median(acceleration), stats.stdev(acceleration))
print(stats.mean(model_year), stats.median(model_year), stats.stdev(model_year))
print(stats.mean(origin), stats.median(origin), stats.stdev(origin))
````
